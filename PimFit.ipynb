```python
import numpy as np
import random
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.stats import chisquare
```

```python
Moment0mod,Event0mod,Event0Uncmod = np.loadtxt(r'MMMomentZ_0to60.txt', unpack=True)
Moment60mod,Event60mod,Event60Uncmod = np.loadtxt(r'MMMomentZ_60to200.txt', unpack=True)
Moment200mod,Event200mod,Event200Uncmod = np.loadtxt(r'MMMomentZ_200to400.txt', unpack=True)
Moment400mod,Event400mod,Event400Uncmod = np.loadtxt(r'MMMomentZ_400to900.txt', unpack=True)
Moment900mod,Event900mod,Event900Uncmod = np.loadtxt(r'MMMomentZ_900to1300.txt', unpack=True)
Moment1300mod,Event1300mod,Event1300Uncmod = np.loadtxt(r'MMMomentZ_1300to2000.txt', unpack=True)

Moment0 = np.zeros(151)
Event0 = np.zeros(151)
Event0Unc = np.zeros(151)

Moment60 = np.zeros(151)
Event60 = np.zeros(151)
Event60Unc = np.zeros(151)

Moment200 = np.zeros(151)
Event200 = np.zeros(151)
Event200Unc = np.zeros(151)

Moment400 = np.zeros(151)
Event400 = np.zeros(151)
Event400Unc = np.zeros(151)

Moment900 = np.zeros(151)
Event900 = np.zeros(151)
Event900Unc = np.zeros(151)

Moment1300 = np.zeros(151)
Event1300 = np.zeros(151)
Event1300Unc = np.zeros(151)

for i in range(Moment0mod.size-1):
    if((i/2) % 2 == 0.0 and i != 600):
        Moment0[int(i/4)] = Moment0mod[i]+Moment0mod[i+1]+Moment0mod[i+2]+Moment0mod[i+3]
        Event0[int(i/4)] = Event0mod[i]+Event0mod[i+1]+Event0mod[i+2]+Event0mod[i+3]
        Event0Unc[int(i/4)] = Event0Uncmod[i]+Event0Uncmod[i+1]+Event0Uncmod[i+2]+Event0Uncmod[i+3]

        Moment60[int(i/4)] = Moment60mod[i]+Moment60mod[i+1]+Moment60mod[i+2]+Moment60mod[i+3]
        Event60[int(i/4)] = Event60mod[i]+Event60mod[i+1]+Event60mod[i+2]+Event60mod[i+3]
        Event60Unc[int(i/4)] = Event60Uncmod[i]+Event60Uncmod[i+1]+Event60Uncmod[i+2]+Event60Uncmod[i+3]

        Moment200[int(i/4)] = Moment200mod[i]+Moment200mod[i+1]+Moment200mod[i+2]+Moment200mod[i+3]
        Event200[int(i/4)] = Event200mod[i]+Event200mod[i+1]+Event200mod[i+2]+Event200mod[i+3]
        Event200Unc[int(i/4)] = Event200Uncmod[i]+Event200Uncmod[i+1]+Event200Uncmod[i+2]+Event200Uncmod[i+3]
        
        Moment400[int(i/4)] = Moment400mod[i]+Moment400mod[i+1]+Moment400mod[i+2]+Moment400mod[i+3]
        Event400[int(i/4)] = Event400mod[i]+Event400mod[i+1]+Event400mod[i+2]+Event400mod[i+3]
        Event400Unc[int(i/4)] = Event400Uncmod[i]+Event400Uncmod[i+1]+Event400Uncmod[i+2]+Event400Uncmod[i+3]
        
        Moment900[int(i/4)] = Moment900mod[i]+Moment900mod[i+1]+Moment900mod[i+2]+Moment900mod[i+3]
        Event900[int(i/4)] = Event900mod[i]+Event900mod[i+1]+Event900mod[i+2]+Event900mod[i+3]
        Event900Unc[int(i/4)] = Event900Uncmod[i]+Event900Uncmod[i+1]+Event900Uncmod[i+2]+Event900Uncmod[i+3]
        
        Moment1300[int(i/4)] = Moment1300mod[i]+Moment1300mod[i+1]+Moment1300mod[i+2]+Moment1300mod[i+3]
        Event1300[int(i/4)] = Event1300mod[i]+Event1300mod[i+1]+Event1300mod[i+2]+Event1300mod[i+3]
        Event1300Unc[int(i/4)] = Event1300Uncmod[i]+Event1300Uncmod[i+1]+Event1300Uncmod[i+2]+Event1300Uncmod[i+3]
        
print(Moment0)
```

```python
def fitfunc(mmz,a,b,c,d,e):                                               
    return (a*np.exp((-(mmz-b)**2)/(2*c**2))) + d/(1+np.exp(-e*mmz))
def gaus(mmz,a,b,c):
    return (a*np.exp((-(mmz-b)**2)/(2*c**2)))
def logist(mmz,d,e):
    return d/(1+np.exp(-e*mmz))

fig0,ax0 = plt.subplots()
fig60,ax60 = plt.subplots()
fig200,ax200 = plt.subplots()
fig400,ax400 = plt.subplots()
fig900,ax900 = plt.subplots()
fig1300,ax1300 = plt.subplots()

initGuess0 = [30,0.5,.6,Event0[149],1]
fitVals0,fitCovar0 = curve_fit(fitfunc, Moment0[0:130], Event0[0:130], p0=initGuess0,sigma=(Event0Unc[0:130]+1),absolute_sigma=True)
fitErr0 = np.sqrt(np.diag(fitCovar0))
ax0.plot(Moment0[0:130],fitfunc(Moment0[0:130],*fitVals0))
ax0.plot(Moment0[0:130],gaus(Moment0[0:130],*fitVals0[0:3]))
ax0.plot(Moment0[0:130],logist(Moment0[0:130],*fitVals0[3:6]))
ax0.scatter(Moment0[0:130],Event0[0:130],color='black',s=1)
ax0.set_title('Missing Z Momentum Counts, 0 < -t < 0.06')
ax0.legend(['Signal and Background','Signal','Background'])
ax0.set_xlabel('Missing Z Momentum (GeV)')
ax0.set_ylabel('Detected Events in t Range')

initGuess60 = [149,0.5,1,Event60[149],1]
fitVals60,fitCovar60 = curve_fit(fitfunc, Moment60[0:130], Event60[0:130], p0=initGuess60,sigma=(Event60Unc[0:130]+1),absolute_sigma=True)
fitErr60 = np.sqrt(np.diag(fitCovar60))
ax60.plot(Moment60[0:130],fitfunc(Moment60[0:130],*fitVals60))
ax60.plot(Moment60[0:130],gaus(Moment60[0:130],*fitVals60[0:3]))
ax60.plot(Moment60[0:130],logist(Moment60[0:130],*fitVals60[3:6]))
ax60.scatter(Moment60[0:130],Event60[0:130],color='black',s=1)
ax60.set_title('Missing Z Momentum Counts, 0.06 < -t < 0.2')
ax60.legend(['Signal and Background','Signal','Background'])
ax60.set_xlabel('Missing Z Momentum (GeV)')
ax60.set_ylabel('Detected Events in t Range')

initGuess200 = [150,0.5,1,Event200[149],1]
fitVals200,fitCovar200 = curve_fit(fitfunc, Moment200[0:130], Event200[0:130], p0=initGuess200,sigma=(Event200Unc[0:130]+1),absolute_sigma=True)
fitErr200 = np.sqrt(np.diag(fitCovar200))
ax200.plot(Moment200[0:130],fitfunc(Moment200[0:130],*fitVals200))
ax200.plot(Moment200[0:130],gaus(Moment200[0:130],*fitVals200[0:3]))
ax200.plot(Moment200[0:130],logist(Moment200[0:130],*fitVals200[3:6]))
ax200.scatter(Moment200[0:130],Event200[0:130],color='black',s=1)
ax200.set_title('Missing Z Momentum Counts, 0.2 < -t < 0.4')
ax200.legend(['Signal and Background','Signal','Background'])
ax200.set_xlabel('Missing Z Momentum (GeV)')
ax200.set_ylabel('Detected Events in t Range')

initGuess400 = [500,0.1,1,Event400[149],0.1]
fitVals400,fitCovar400 = curve_fit(fitfunc, Moment400[0:125], Event400[0:125], p0=initGuess400,sigma=(Event400Unc[0:125]+1),absolute_sigma=True)
fitErr400 = np.sqrt(np.diag(fitCovar400))
ax400.plot(Moment400[0:125],fitfunc(Moment400[0:125],*fitVals400))
ax400.plot(Moment400[0:125],gaus(Moment400[0:125],*fitVals400[0:3]))
ax400.plot(Moment400[0:125],logist(Moment400[0:125],*fitVals400[3:6]))
ax400.scatter(Moment400[0:125],Event400[0:125],color='black',s=1)
ax400.set_title('Missing Z Momentum Counts, 0.4 < -t < 0.9')
ax400.legend(['Signal and Background','Signal','Background'])
ax400.set_xlabel('Missing Z Momentum (GeV)')
ax400.set_ylabel('Detected Events in t Range')

initGuess900 = [500,0.1,1,Event60[149],0.1]
fitVals900,fitCovar900 = curve_fit(fitfunc, Moment900[0:125], Event900[0:125], p0=initGuess900,sigma=(Event900Unc[0:125]+1),absolute_sigma=True)
fitErr900 = np.sqrt(np.diag(fitCovar900))
ax900.plot(Moment900[0:125],fitfunc(Moment900[0:125],*fitVals900))
ax900.plot(Moment900[0:125],gaus(Moment900[0:125],*fitVals900[0:3]))
ax900.plot(Moment900[0:125],logist(Moment900[0:125],*fitVals900[3:6]))
ax900.scatter(Moment900[0:125],Event900[0:125],color='black',s=1)
ax900.set_title('Missing Z Momentum Counts, 0.9 < -t < 1.3')
ax900.legend(['Signal and Background','Signal','Background'])
ax900.set_xlabel('Missing Z Momentum (GeV)')
ax900.set_ylabel('Detected Events in t Range')

initGuess1300 = [500,0.1,1,Event60[149],0.1]
fitVals1300,fitCovar1300 = curve_fit(fitfunc, Moment1300[0:125], Event1300[0:125], p0=initGuess1300,sigma=(Event1300Unc[0:125]+1),absolute_sigma=True)
fitErr1300 = np.sqrt(np.diag(fitCovar1300))
ax1300.plot(Moment1300[0:125],fitfunc(Moment1300[0:125],*fitVals1300))
ax1300.plot(Moment1300[0:125],gaus(Moment1300[0:125],*fitVals1300[0:3]))
ax1300.plot(Moment1300[0:125],logist(Moment1300[0:125],*fitVals1300[3:6]))
ax1300.scatter(Moment1300[0:125],Event1300[0:125],color='black',s=1)
ax1300.set_title('Missing Z Momentum Counts, 1.3 < -t < 2.0')
ax1300.legend(['Signal and Background','Signal','Background'])
ax1300.set_xlabel('Missing Z Momentum (GeV)')
ax1300.set_ylabel('Detected Events in t Range')

plt.show()
```
